# Predator-Prey simulation

### Set up Python libraries

As usual, run the code cell below to import the relevant Python libraries

```python
# Set-up Python libraries - you need to run this but you don't need to change it
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import pandas 
import seaborn as sns
sns.set_theme()
```

```python
days = 10000
rabbit_count = np.empty(days+1)
fox_count = np.empty(days+1)
time = np.empty(days+1)
rabbit_count[0]=1
fox_count[0]=0.5
time[t]=0

a = 0.7 # rabbit fertility rate
b = 0.5 # chance of any given rabbit being eaten by any given fox
c = 0.3 # fox death rate (if there were no rabbits to eat)
e = 0.2 # increase in chance of fox surviving if it eats a rabbit

for t in np.arange(start=1, stop=days+1, step=1):
    
    r = rabbit_count[t-1]
    f = fox_count[t-1]
    
    r = r + (a*r - b*r*f)*(1/365)
    f = f + (-c*f + e*r*f)*(1/365)
    
    rabbit_count[t]=r
    fox_count[t]=f
    time[t]=time[t-1]+1/365

plt.plot(fox_count,'r')
plt.plot(rabbit_count,'b')
```

```python

# model parameters
a = 0.7; b = 0.5; c = 0.3;  e = 0.2
dt = 0.001; max_time = 100

# initial time and populations
t = 0; x = 100; y = 50

# empty lists in which to store time and populations
t_list = []; x_list = []; y_list = []

# initialize lists
t_list.append(t); x_list.append(x); y_list.append(y)

while t < max_time:
    # calc new values for t, x, y
    t = t + dt
    x = x + (a*x - b*x*y)*dt
    y = y + (-c*y + e*x*y)*dt

    # store new values in lists
    t_list.append(t)
    x_list.append(x)
    y_list.append(y)

# Plot the results    
p = plt.plot(t_list, x_list, 'r', t_list, y_list, 'g', linewidth = 2)

```
