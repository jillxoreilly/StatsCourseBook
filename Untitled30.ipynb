### Import and view the data

```python
movies=pandas.read_csv('data/MovieProfits.csv')
movies
```

```python
sns.scatterplot(data=movies, x='Budget ($M)', y='Critic Score (Rotten Tomatoes)')
movies['Budget ($M)'].corr(movies['Critic Score (Rotten Tomatoes)'], method='spearman')
```

```python
movies.sort_values(by='US Gross ($M)')
```

```python
movies['Profit per dollar']=movies['Profit ($M)']/movies['Budget ($M)']
movies
```

```python
sns.histplot(data=movies, y='Profit per dollar', x='Budget ($M)')
```

```python
sns.scatterplot(data=movies, x='Critic Score (Rotten Tomatoes)', y='US Gross ($M)')
```

```python
sns.violinplot(data=movies, x='Year', y='Budget ($M)')
```
